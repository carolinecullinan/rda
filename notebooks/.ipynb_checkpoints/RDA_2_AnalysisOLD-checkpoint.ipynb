```python
# https://www.dotactiv.com/blog/how-to-use-demographic-clustering
# https://www.r-bloggers.com/2020/07/demographic-modeling-cluster-analysis/


# https://crunchingthedata.com/when-to-use-k-means-clustering/#:~:text=K%2Dmeans%20clustering%20is%20generally,observations%20that%20share%20similar%20characteristics.


# https://towardsdatascience.com/clustering-datasets-having-both-numerical-and-categorical-variables-ed91cdca0677
# x: treatment days ('treatment_days')
# y: success rate (absolute_value('change_freq_relative')


#https://medium.com/analytics-vidhya/clustering-on-mixed-data-types-in-python-7c22b3898086
```

```python
# normalize continuous features in data to ensure that one feature is not interpreted
    ## as being more important than others
from sklearn import preprocessing
data_norm = data.copy()
scaler = preprocessing.MinMaxScaler()
data_norm[['age','treatment_days', 'start_freq_relative', 'exit_freq_relative', 
          'change_freq_relative']] = scaler.fit_transform(data_norm[['age',
          'treatment_days', 'start_freq_relative', 'exit_freq_relative', 
          'change_freq_relative']])
```

```python
# Note: In 2009, Cao et al. [24] introduced the initialization method that works by the distance between samples and the density of samples. Their method selects a sample with the highest mean density as the initial center for a cluster
```

```python
data.head()
```

```python
# implement K-prototypes algorithm to cluster data (3 clusters and Cao initialization)
from kmodes.kprototypes import KPrototypes
kproto = KPrototypes(n_clusters=3, init='Cao')
clusters = kproto.fit_predict(data_norm, categorical=[1, 2, 3, 4, 5, 6, 7, 12]) #columns that are categorical = 1, 2, 3, 4, 5, 6, 7, 12 
#join data with labels 
labels = pd.DataFrame(clusters)
labeledData = pd.concat((data,labels),axis=1)
labeledData = labeledData.rename({0:'labels'},axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

labeledData['Constant'] = 0 #dummy feature for plotting

f, axes = plt.subplots(1, 4, figsize=(25, 7), sharex=False)
f.subplots_adjust(hspace=0.2, wspace=0.7)

for i in range(4):
    col = labeledData.columns[i]
    if i < 2:    
        sns.catplot(x=col, y='labels', kind="swarm", hue='labels', data=labeledData,ax=axes[i])
    else:
        ax = sns.swarmplot(x=labeledData['Constant'],y=labeledData[col].values,hue=labeledData['labels'],ax=axes[i])
        ax.set_title(col)
        
plt.close(2)
plt.close(3)
plt.show()
```
